# Media Cloud

## Fetch All Sources

```python
MEDIA_CLOUD_KEY = os.environ['MEDIA_CLOUD_API_KEY']
```

```python
media = []
last_media_id = 0
rows  = 1000
while True:
    params = { 'last_media_id': last_media_id, 'rows': rows, 'key': MEDIA_CLOUD_KEY }
    print("last_media_id:{} rows:{}".format( last_media_id, rows))
    r = requests.get( 'https://api.mediacloud.org/api/v2/media/list', params = params, headers = { 'Accept': 'application/json'} )
    data = r.json()

    if len(data) == 0:
        break

    last_media_id = data[-1]['media_id']
    media.extend( data )
```

```python
mc_raw = pd.DataFrame.from_records(media)
```

```python
mc_raw.to_csv('../data/raw/media_cloud_all_sources.csv')
```

```python
with open('../data/raw/media_cloud_all_sources.pickle', 'wb') as write_file:
    pickle.dump(mc_raw, write_file)
```

## Process

```python
mc_df = mc_raw
```

```python
def get_domain(url):
    try:
        return tld.get_tld(url, as_object=True).domain
    except KeyboardInterrupt:
        raise
    except:
        print(url)
#         print(e)
        return np.nan
```

```python
mc_df['tld'] = mc_df.url.map(get_domain)
```

```python
mc_df['tld_dup_count'] = mc_df.groupby('tld').url.transform('count')
```

```python
mc_df = mc_df.drop_duplicates(subset='tld', keep='first')
```

## Reddit Coverage

```python
posts_df = pd.DataFrame.from_records(
    db.climate_submissions.find(
        {},
        {'score': 1, 'num_comments': 1, 'url': 1, 'created_utc': 1}
    )
).sort_values('num_comments', ascending=False).drop_duplicates(subset=['url'], keep='first')
posts_df['tld'] = posts_df.url.map(get_domain)
mereged_df = posts_df.merge(mc_df, how='left', on='tld')
```
