# Crowd Tangle

```python
import requests
```

```python
def get_crowd_tangle(link, platforms, ct_token):
#     platforms_query = ','.join(platforms)
#     ct_links_req = f'https://api.crowdtangle.com/ce/links?link={link}&platforms={platforms_query}&token={ct_token}&version=3.0.27'
#     links_resp = requests.get(ct_links_req)
    
    ct_inter_req = f'https://api.crowdtangle.com/ce/interactions?link={link}&token={ct_token}'
    inter_resp = requests.get(ct_inter_req)
#     return links_resp.json(), inter_resp.json()
    return inter_resp.json()
```

```python
links = get_crowd_tangle(
    link='https://www.cnn.com/2020/03/06/africa/agriculture-regenerative-farming-climate-crisis-intl',
    platforms=['facebook'],
    ct_token=os.environ['CROWD_TANGLE_API_TOKEN']
)
```
