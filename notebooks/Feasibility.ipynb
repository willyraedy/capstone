```python
import requests
```

```python
climate_resp = requests.get('https://api.pushshift.io/reddit/search/submission/?subreddit=worldnews&limit=500&q=climate')
```

```python
test_urls = [x['url'] for x in climate_resp.json()['data']]
```

```python
def create_fetch_text():
    count = 0
    success = 0
    fail = 0
    def fetch_text(url):
        diff_test_resp = requests.get(f'https://api.diffbot.com/v3/article?url={url}&token=9a1176e979dfb2ce21c481688f8e5529')
        nonlocal count
        nonlocal success
        nonlocal fail
        count += 1
        if count % 10 == 0:
            print('Count: ', count)
            print('Succes:', success)
            print('Fail:', fail)
        try:
            text = diff_test_resp.json()['objects'][0]['text']
            if len(text) > 0:
                success += 1
            else:
                fail += 1
            return text
        except Exception as e:
            fail += 1
            return e
    return fetch_text
```

```python
fetch_with_closure = create_fetch_text()
diff_results = [fetch_with_closure(x) for x in test_urls]
```
